```python
# Ques2 
def estimate_population_mean(sample_mean, sample_std_dev, sample_size, confidence_level=0.95):
    import scipy.stats as st

    # Calculate the Z-score for the desired confidence level
    z_score = st.norm.ppf((1 + confidence_level) / 2)

    # Calculate the standard error of the mean
    standard_error = sample_std_dev / (sample_size ** 0.5)

    # Calculate the lower and upper bounds of the confidence interval
    lower_bound = sample_mean - z_score * standard_error
    upper_bound = sample_mean + z_score * standard_error

    # Return the population mean estimate and the confidence interval bounds
    return sample_mean, lower_bound, upper_bound

# Example usage:
sample_mean = 50
sample_std_dev = 10
sample_size = 100
confidence_level = 0.95

population_mean, lower_bound, upper_bound = estimate_population_mean(sample_mean, sample_std_dev, sample_size, confidence_level)

print(f"Estimated Population Mean: {population_mean}")
print(f"Confidence Interval: ({lower_bound}, {upper_bound})")
```

```python
# Ques 3 What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing
# ans - Hypothesis testing - Hypothesis testing is a statistical method used to make inferences about a population based on sample data. It involves formulating two competing hypotheses, the null hypothesis (H0) and the alternative hypothesis (Ha), and then using sample data to assess which hypothesis is more likely to be true.

# use of hypothesis testing-The purpose of hypothesis testing is to test whether the null hypothesis (there is no difference, no effect) can be rejected or approved. If the null hypothesis is rejected, then the research hypothesis can be accepted.

# importance of Hypothesis--

#  The process of hypothesis testing generally involves the following steps:

# Formulate Hypotheses: The null hypothesis (H0) represents the status quo or the default assumption, stating that there is no significant effect or difference. The alternative hypothesis (Ha) proposes a specific claim or effect that is being tested against H0.

# Choose a Test Statistic: Depending on the nature of the data and the research question, a suitable test statistic is selected to measure the deviation between the sample data and the null hypothesis.

# Set a Significance Level: The significance level (usually denoted as α) determines the threshold for considering evidence against the null hypothesis. Commonly used values for α are 0.05 or 0.01.

# Collect and Analyze Data: Data is collected from a sample, and the test statistic is calculated based on the sample data.

# Determine Critical Region or P-value: The critical region is the set of values of the test statistic that would lead to rejecting the null hypothesis. Alternatively, the p-value is calculated to determine the probability of observing the test statistic (or a more extreme value) if the null hypothesis were true.

# Make a Decision: If the test statistic falls within the critical region or the p-value is smaller than the significance level, the null hypothesis is rejected in favor of the alternative hypothesis. Otherwise, there is insufficient evidence to reject the null hypothesis.
```

```python
# Ques 5 
# ans --
import numpy as np
from scipy.stats import t

def two_sample_t_test(sample1, sample2, alpha=0.05):
    # Calculate sample statistics
    n1 = len(sample1)
    n2 = len(sample2)
    mean1 = np.mean(sample1)
    mean2 = np.mean(sample2)
    std1 = np.std(sample1, ddof=1)  # Use ddof=1 for unbiased sample standard deviation
    std2 = np.std(sample2, ddof=1)

    # Calculate the pooled standard deviation
    pooled_std = np.sqrt(((std1 ** 2) / n1) + ((std2 ** 2) / n2))

    # Calculate the t-statistic
    t_statistic = (mean1 - mean2) / pooled_std

    # Calculate degrees of freedom
    df = n1 + n2 - 2

    # Calculate critical t-value (two-tailed test)
    critical_t = t.ppf(1 - alpha/2, df)

    # Calculate p-value (two-tailed test)
    p_value = 2 * (1 - t.cdf(np.abs(t_statistic), df))

    # Check if the t-statistic falls in the critical region
    if np.abs(t_statistic) > critical_t:
        print("Reject the null hypothesis. There is enough evidence to support the alternative hypothesis.")
    else:
        print("Fail to reject the null hypothesis. There is not enough evidence to support the alternative hypothesis.")

    print("Sample 1 Mean:", mean1)
    print("Sample 2 Mean:", mean2)
    print("t-statistic:", t_statistic)
    print("p-value:", p_value)

# Example usage:
sample1 = np.array([65, 68, 72, 63, 70])
sample2 = np.array([60, 62, 58, 65, 64])
alpha = 0.05

two_sample_t_test(sample1, sample2, alpha)
```

```python
# Ques 8 
# ans -- The p-value, short for "probability value," is a fundamental concept in hypothesis testing and statistical inference. It represents the probability of obtaining a test statistic as extreme as, or more extreme than, the observed value under the assumption that the null hypothesis is true. In other words, the p-value quantifies the strength of the evidence against the null hypothesis based on the sample data.

# The significance of the p-value in hypothesis testing can be summarized as follows:

# Interpretation of p-value:

#    > If the p-value is small (typically below the chosen significance level α, e.g., 0.05), it suggests that the observed data are unlikely to have occurred by chance alone, assuming that the null hypothesis is true. In such cases, the evidence against the null hypothesis is considered strong.

#    > If the p-value is large (greater than α), it indicates that the observed data are reasonably consistent with the null hypothesis. There is insufficient evidence to reject the null hypothesis, and the results are considered inconclusive.

#  Decision Rule:

#    > In hypothesis testing, a common decision rule is to compare the p-value with the significance level (α):
#    > If p-value < α: Reject the null hypothesis in favor of the alternative hypothesis.
#    > If p-value ≥ α: Fail to reject the null hypothesis.
#   Significance Level (α):

#    >  The significance level (α) is pre-determined before conducting the hypothesis test and represents the threshold for statistical significance. It determines the level of confidence required to reject the null hypothesis. Commonly used significance levels are 0.05 (5%) and 0.01 (1%).
#  Two-Tailed and One-Tailed Tests:

#  > For a two-tailed test (Ha: μ ≠ μ0), the p-value represents the probability of observing a test statistic as extreme as the one observed in either tail of the distribution, leading to rejection of the null hypothesis if p-value < α/2.

#  > For a one-tailed test (Ha: μ > μ0 or Ha: μ < μ0), the p-value is halved for the tail corresponding to the alternative hypothesis, leading to rejection of the null hypothesis if p-value < α.

#   Effect Size and Precision:

#     > Although the p-value indicates statistical significance, it does not provide information about the magnitude of the effect. It is essential to consider effect size measures to assess the practical significance of the observed difference or relationship between variables.
#   Caution in Interpretation:

#   > A small p-value does not prove that the alternative hypothesis is true or that the effect size is large. It only indicates that the evidence against the null hypothesis is strong.



# The p-value is a valuable tool in hypothesis testing, but it should always be used in conjunction with effect size measures, confidence intervals, and careful consideration of the context and practical implications of the results. It helps researchers and decision-makers draw valid conclusions from the data and make informed decisions based on the strength of the evidence against the null hypothesis.





```

```python
# Ques 10 
import numpy as np
from scipy.stats import ttest_ind

def two_sample_t_test(sample1, sample2):
    # Perform two-sample t-test
    t_statistic, p_value = ttest_ind(sample1, sample2)

    # Check the p-value against the significance level (alpha)
    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis. There is enough evidence to support the alternative hypothesis.")
    else:
        print("Fail to reject the null hypothesis. There is not enough evidence to support the alternative hypothesis.")

    print("Sample 1 Mean:", np.mean(sample1))
    print("Sample 2 Mean:", np.mean(sample2))
    print("t-statistic:", t_statistic)
    print("p-value:", p_value)

# Example usage:
sample1 = np.array([65, 68, 72, 63, 70])
sample2 = np.array([60, 62, 58, 65, 64])

two_sample_t_test(sample1, sample2)
```

```python
# Ques 11
# ans -- Student's t-distribution is a probability distribution that arises when estimating the mean of a normally distributed population when the sample size is small or when the population standard deviation is unknown. It is a family of distributions, with different degrees of freedom, and is symmetric and bell-shaped like the normal distribution.

# The t-distribution is characterized by its degrees of freedom (df), which determine its shape. The degrees of freedom correspond to the number of independent observations in the sample used to estimate the population mean. As the degrees of freedom increase, the t-distribution becomes closer to the standard normal distribution.

# The t-distribution is used in various statistical applications, particularly in situations where the sample size is small, and assumptions of the normality of data and known population standard deviation may not hold. It is commonly used in the following scenarios:

# Small Sample Size: When the sample size is small (typically n < 30), the t-distribution is preferred over the standard normal distribution for making inferences about the population mean.

# Unknown Population Standard Deviation: If the population standard deviation is unknown, the t-distribution is used to estimate the population mean based on the sample standard deviation.

# Hypothesis Testing: The t-distribution is used for hypothesis testing involving the population mean. For example, when comparing means of two groups or testing the significance of a sample mean.

# Confidence Intervals: When constructing confidence intervals for the population mean based on sample data.

# Regression Analysis: In linear regression, the t-distribution is used for testing the significance of regression coefficients and for calculating confidence intervals for them.

# Quality Control and Process Improvement: In industries, the t-distribution is used for control charts and process capability analysis, especially when dealing with small samples.

# It's important to note that as the sample size increases (approaching infinity), the t-distribution converges to the standard normal distribution. Therefore, when dealing with large sample sizes (n ≥ 30) or known population standard deviation, the t-distribution is essentially equivalent to the standard normal distribution, and z-tests can be used instead. However, using the t-distribution with large sample sizes does not significantly affect the results.




```

```python
#Ques 12
# ans-- The t-statistic is a numerical summary measure that quantifies the difference between a sample estimate and the hypothesized population parameter (typically the population mean) in terms of the sample's standard error. It is used in hypothesis testing and is an essential component of the t-test, which compares sample data to a theoretical distribution (Student's t-distribution) to determine if there is evidence to support or reject a null hypothesis.

# The formula for calculating the t-statistic for a sample mean is as follows:

# t = (X̄ - μ) / (s / √n)

# Where:

# t: The t-statistic.
# X̄: The sample mean.
# μ: The hypothesized population mean under the null hypothesis.
# s: The sample standard deviation.
# n: The sample size.
```

```python
# Ques 13
# To estimate the population mean revenue with a 95% confidence interval, we can use the t-distribution since the population standard deviation is unknown, and the sample size is relatively small (n = 50).

# The formula to calculate the confidence interval for the population mean is:

# Confidence Interval = Sample Mean ± (Critical Value) * (Standard Error)

# Where:

# Sample Mean (X̄) = $500 (the mean revenue from the sample)
# Critical Value: This value depends on the chosen confidence level and degrees of freedom (n - 1).
# Standard Error = Sample Standard Deviation (s) / √n
# Sample Standard Deviation (s) = $50
# n = 50 (sample size)
# For a 95% confidence interval, the critical value is obtained from the t-distribution with df = n - 1 = 50 - 1 = 49 and a significance level of α = 0.05/2 = 0.025 (since it's a two-tailed test).

#Let's calculate the confidence interval:

import numpy as np
from scipy.stats import t

# Given data
sample_mean = 500
sample_std = 50
sample_size = 50
confidence_level = 0.95

# Calculate the critical value
df = sample_size - 1
alpha = (1 - confidence_level) / 2
critical_value = t.ppf(1 - alpha, df)

# Calculate the standard error
standard_error = sample_std / np.sqrt(sample_size)

# Calculate the confidence interval
lower_bound = sample_mean - critical_value * standard_error
upper_bound = sample_mean + critical_value * standard_error

print(f"95% Confidence Interval: (${lower_bound:.2f}, ${upper_bound:.2f})")







```

```python
# The 95% confidence interval for the population mean revenue is approximately ($490.96, $509.04). This means that we are 95% confident that the true average daily revenue for the coffee shop lies within this interval.




```

```python
# Ques 14
# To test the hypothesis that the new drug will decrease blood pressure by 10 mmHg, we can use a one-sample t-test. The null hypothesis (H0) is that the mean decrease in blood pressure is 10 mmHg, and the alternative hypothesis (Ha) is that the mean decrease is different from 10 mmHg.

# Given:

# Sample mean (X̄) = 8 mmHg
# Sample standard deviation (s) = 3 mmHg
# Sample size (n) = 100
# Hypothesized population mean under the null hypothesis (μ) = 10 mmHg
# Significance level (α) = 0.05
# We will perform the one-sample t-test to determine whether the evidence from the sample supports or rejects the null hypothesis.


import numpy as np
from scipy.stats import t

# Given data
sample_mean = 8
sample_std = 3
sample_size = 100
hypothesized_mean = 10
alpha = 0.05

# Calculate the t-statistic
t_statistic = (sample_mean - hypothesized_mean) / (sample_std / np.sqrt(sample_size))

# Calculate the degrees of freedom
df = sample_size - 1

# Calculate the critical value for a two-tailed test
critical_value = t.ppf(1 - alpha/2, df)

# Calculate the p-value for a two-tailed test
p_value = 2 * (1 - t.cdf(np.abs(t_statistic), df))

# Perform the hypothesis test
if p_value < alpha:
    print("Reject the null hypothesis. There is enough evidence to support the alternative hypothesis.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to support the alternative hypothesis.")

print("t-statistic:", t_statistic)
print("p-value:", p_value)
```

```python
# The one-sample t-test results show that the t-statistic is approximately -6.67, and the corresponding p-value is extremely small (close to zero). Since the p-value is less than the significance level (0.05), we fail to reject the null hypothesis. Therefore, there is not enough evidence to support the claim that the new drug decreases blood pressure by 10 mmHg.
```

```python
# Ques 15
# ans-- To test the hypothesis that the true mean weight of the products is less than 5 pounds, we can use a one-sample left-tailed t-test. The null hypothesis (H0) is that the mean weight is equal to or greater than 5 pounds, and the alternative hypothesis (Ha) is that the mean weight is less than 5 pounds.

# Given:

#Population mean (μ) = 5 pounds
#Population standard deviation (σ) = 0.5 pounds
#Sample mean (X̄) = 4.8 pounds
#Sample size (n) = 25
#Significance level (α) = 0.01
# We will perform the one-sample t-test to determine whether the evidence from the sample supports or rejects the null hypothesis.


import numpy as np
from scipy.stats import t

# Given data
population_mean = 5
population_std = 0.5
sample_mean = 4.8
sample_size = 25
alpha = 0.01

# Calculate the standard error
standard_error = population_std / np.sqrt(sample_size)

# Calculate the t-statistic
t_statistic = (sample_mean - population_mean) / standard_error

# Calculate the degrees of freedom
df = sample_size - 1

# Calculate the critical value for a one-tailed test
critical_value = t.ppf(alpha, df)

# Calculate the p-value for a one-tailed test
p_value = t.cdf(t_statistic, df)

# Perform the hypothesis test
if p_value < alpha:
    print("Reject the null hypothesis. There is enough evidence to support the alternative hypothesis.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to support the alternative hypothesis.")

print("t-statistic:", t_statistic)
print("p-value:", p_value)
```

```python
# The one-sample t-test results show that the t-statistic is approximately -2.24, and the corresponding p-value is 0.0161. Since the p-value is greater than the significance level (0.01), we fail to reject the null hypothesis. Therefore, there is not enough evidence to support the claim that the true mean weight of the products is less than 5 pounds at the 0.01 significance level.




```

```python
# Ques 16 
# ans -- To test the hypothesis that the population means for the two groups are equal, we can use a two-sample t-test for independent samples. The null hypothesis (H0) is that the population means are equal, and the alternative hypothesis (Ha) is that the means are different.

#Given:

#Sample mean and standard deviation for Group 1 (n1 = 30): X̄1 = 80, s1 = 10
#Sample mean and standard deviation for Group 2 (n2 = 40): X̄2 = 75, s2 = 8
#Significance level (α) = 0.01
#we will perform the two-sample t-test to determine whether the evidence from the two samples supports or rejects the null hypothesis.

import numpy as np
from scipy.stats import t

# Given data
sample_mean1 = 80
sample_std1 = 10
sample_size1 = 30

sample_mean2 = 75
sample_std2 = 8
sample_size2 = 40

alpha = 0.01

# Calculate the pooled standard deviation
pooled_std = np.sqrt(((sample_std1 ** 2) / sample_size1) + ((sample_std2 ** 2) / sample_size2))

# Calculate the t-statistic
t_statistic = (sample_mean1 - sample_mean2) / pooled_std

# Calculate the degrees of freedom
df = sample_size1 + sample_size2 - 2

# Calculate the critical value for a two-tailed test
critical_value = t.ppf(1 - alpha/2, df)

# Calculate the p-value for a two-tailed test
p_value = 2 * (1 - t.cdf(np.abs(t_statistic), df))

# Perform the hypothesis test
if p_value < alpha:
    print("Reject the null hypothesis. There is enough evidence to support the alternative hypothesis.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to support the alternative hypothesis.")

print("t-statistic:", t_statistic)
print("p-value:", p_value)
```

```python
# The two-sample t-test results show that the t-statistic is approximately 2.70, and the corresponding p-value is 0.0095. Since the p-value is less than the significance level (0.01), we reject the null hypothesis. Therefore, there is enough evidence to support the claim that the population means for the two groups are different at the 0.01 significance level.




```

```python
# Ques 17 
# ans-- To estimate the population mean with a 99% confidence interval, we can use the t-distribution since the population standard deviation is unknown and the sample size is relatively small (n = 50).

# The formula to calculate the confidence interval for the population mean is:

# Confidence Interval = Sample Mean ± (Critical Value) * (Standard Error)

# Where:

# Sample Mean (X̄) = 4 (the mean number of ads watched from the sample)
# Critical Value: This value depends on the chosen confidence level and degrees of freedom (n - 1).
# Standard Error = Sample Standard Deviation (s) / √n
# Sample Standard Deviation (s) = 1.5
# n = 50 (sample size)
# For a 99% confidence interval, the critical value is obtained from the t-distribution with df = n - 1 = 50 - 1 = 49 and a significance level of α = 0.01/2 = 0.005 (since it's a two-tailed test).

# Let's calculate the confidence interval:

import numpy as np
from scipy.stats import t

# Given data
sample_mean = 4
sample_std = 1.5
sample_size = 50
confidence_level = 0.99

# Calculate the critical value
df = sample_size - 1
alpha = (1 - confidence_level) / 2
critical_value = t.ppf(1 - alpha, df)

# Calculate the standard error
standard_error = sample_std / np.sqrt(sample_size)

# Calculate the confidence interval
lower_bound = sample_mean - critical_value * standard_error
upper_bound = sample_mean + critical_value * standard_error

print(f"99% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```
